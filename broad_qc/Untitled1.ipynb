```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the updated Excel file
file_path = '/gale/netapp/seq11/illumina_runs/qc_metrics/BICAN_total_reads_pf.xlsx'  # Update with your file path
df_sheet2 = pd.read_excel(file_path, sheet_name='Sheet2')

# Define the pairs of SALK pools to color the same
color_pairs = {
    (52, 54): 'pink',
    (56, 57): 'orange',
    (59, 60): 'darkblue',
    (58, 61): 'green',
    (62, 63): 'blue',
    (64, 65): 'purple',
    (66, 67): 'red',
    (68, 69): 'cyan',
    (70, 71): 'brown'
}

# Create a color list based on the specified pairs
colors = []
for pool in df_sheet2['SALK pool #']:
    color = 'skyblue'  # default color
    for pair, col in color_pairs.items():
        if pool in pair:
            color = col
            break
    colors.append(color)

# Plot the bar chart with the swapped colors for red and pink pairs
plt.figure(figsize=(12, 8))
plt.bar(df_sheet2.index, df_sheet2['Total. Reads PF'] / 1e9, color=colors, width=0.8)
plt.xlabel('SALK pool #')
plt.ylabel('Total Reads PF (Billions)')
plt.title('Total Reads PF per SALK Pool')

# Rotate x-axis labels 90 degrees for better readability
plt.xticks(df_sheet2.index, labels=df_sheet2['SALK pool #'].astype(int), rotation=90)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  # Ensure everything fits without overlap

plt.show()
```
