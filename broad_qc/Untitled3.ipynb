```python
#!/usr/bin/env python3

import pandas as pd

def calculate_read_proportion(df, pool_pairs):
    """
    Calculate and add the read proportions and run pairs to the DataFrame for multiple pairs of pools.

    This function takes a DataFrame containing read counts for different pools
    and a list of tuples, each containing two pool names. It calculates the ratio of total reads
    between each specified pair of pools and adds the result to the DataFrame
    in the 'Read Proportion' column for each pool. It also adds a 'Run Pair'
    column to indicate which pools are paired together for sequencing, reflecting
    the order of numerator and denominator in the pair.

    Parameters:
    df (pd.DataFrame): A DataFrame containing at least two columns:
                       'Pool' (str) - the name of the pool
                       'Total Reads' (int) - the total number of reads for the pool.
    pool_pairs (list of tuples): A list of tuples, each containing the names of two pools as strings.
                                 Example: [('SALK082', 'SALK084'), ('SALK083', 'SALK085')]

    Returns:
    pd.DataFrame: The updated DataFrame with the 'Read Proportion' and 'Run Pair' columns added.

    Raises:
    ValueError: If any of the pools in the pairs are not found in the DataFrame.

    Example:
    >>> summary_df = pd.DataFrame({
            'Pool': ['SALK082', 'SALK084', 'SALK083', 'SALK085'],
            'Total Reads': [610297, 387352, 500000, 400000]
        })
    >>> updated_df = calculate_read_proportion(summary_df, [('SALK082', 'SALK084'), ('SALK083', 'SALK085')])
    >>> print(updated_df)
           Pool  Total Reads  Read Proportion         Run Pair
    0  SALK082       610297          0.634694  SALK082_SALK084
    1  SALK084       387352          1.575562  SALK084_SALK082
    2  SALK083       500000          0.800000  SALK083_SALK085
    3  SALK085       400000          1.250000  SALK085_SALK083
    """
    # Validate the presence of required columns
    required_columns = ['Pool', 'Total Reads']
    if not all(column in df.columns for column in required_columns):
        raise ValueError(f"The DataFrame must contain the following columns: {required_columns}")
    
    # Validate the presence of specified pools
    for pair in pool_pairs:
        for pool in pair:
            if pool not in df['Pool'].values:
                raise ValueError(f"Pool '{pool}' not found in the DataFrame.")
    
    # Initialize 'Read Proportion' and 'Run Pair' columns
    df['Read Proportion'] = None
    df['Run Pair'] = None
    
    # Calculate and add the proportions for each pair
    for pools in pool_pairs:
        # Extract the total reads for the given pools
        reads_1 = df.loc[df['Pool'] == pools[0], 'Total Reads'].values[0]
        reads_2 = df.loc[df['Pool'] == pools[1], 'Total Reads'].values[0]
        
        # Calculate the proportions
        proportion_1 = reads_2 / reads_1  # Proportion for pools[0]
        proportion_2 = reads_1 / reads_2  # Proportion for pools[1]
        
        # Add the proportions to the respective rows
        df.loc[df['Pool'] == pools[0], 'Read Proportion'] = proportion_1
        df.loc[df['Pool'] == pools[1], 'Read Proportion'] = proportion_2
        
        # Add the run pair to the respective rows
        run_pair_1 = f"{pools[0]}_{pools[1]}"
        run_pair_2 = f"{pools[1]}_{pools[0]}"
        df.loc[df['Pool'] == pools[0], 'Run Pair'] = run_pair_1
        df.loc[df['Pool'] == pools[1], 'Run Pair'] = run_pair_2
    
    return df

# Create the DataFrame
summary_df = pd.DataFrame({
    'Pool': ['SALK082', 'SALK084', 'SALK083', 'SALK085'],
    'Total Reads': [610297, 387352, 500000, 400000]
})

# Test the function
updated_df = calculate_read_proportion(summary_df, [('SALK082', 'SALK084'), ('SALK083', 'SALK085')])
print(updated_df)
```

```python
def calculate_read_proportion(pools_data):
    """
    Calculate read proportions for a pair of pools.
    
    Parameters:
    pools_data (list of tuples): Each tuple contains ('pool_name', total_reads) for two pools.
    
    Returns:
    list of tuples: Each tuple contains (pool_name, read_proportion, run_pair).
    """
    pool1, pool2 = pools_data
    name1, reads1 = pool1
    name2, reads2 = pool2

    proportion1 = reads2 / reads1
    proportion2 = reads1 / reads2

    return [
        (name1, proportion1, f"{name1}_{name2}"),
        (name2, proportion2, f"{name2}_{name1}")
    ]
```

```python
import sqlite3

def update_database_with_proportions(db_name, pools_data):
    """
    Update the SQLite database with read proportions and run pairs.
    
    Parameters:
    db_name (str): Name of the SQLite database file.
    pools_data (list of tuples): Each tuple contains ('pool_name', total_reads) for two pools.
    """
    # Calculate read proportions
    proportions = calculate_read_proportion(pools_data)
    
    # Connect to SQLite database
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()

    # Update the Pools table with read proportions and run pairs
    for pool_name, read_proportion, run_pair in proportions:
        cursor.execute("""
            UPDATE Pools
            SET read_proportion = ?, run_pair = ?
            WHERE name = ?;
        """, (read_proportion, run_pair, pool_name))
    
    # Commit the changes and close the connection
    conn.commit()
    conn.close()

# Example usage
pools_data = [('SALK082', 610297), ('SALK084', 387352)]
update_database_with_proportions('example.db', pools_data)
```

```python
import pandas as pd

def calculate_read_proportion(df, pool_pairs):
    """
    Calculate and add the read proportions and run pairs to the DataFrame for multiple pairs of pools.

    This function takes a DataFrame containing read counts for different pools
    and a list of tuples, each containing two pool names. It calculates the ratio of total reads
    between each specified pair of pools and adds the result to the DataFrame
    in the 'Read Proportion' column for each pool. It also adds a 'Run Pair'
    column to indicate which pools are paired together for sequencing, reflecting
    the order of numerator and denominator in the pair.

    Parameters:
    df (pd.DataFrame): A DataFrame containing at least two columns:
                       'Pool' (str) - the name of the pool
                       'Total Reads' (int) - the total number of reads for the pool.
    pool_pairs (list of tuples): A list of tuples, each containing the names of two pools as strings.
                                 Example: [('SALK082', 'SALK084'), ('SALK083', 'SALK085')]

    Returns:
    pd.DataFrame: The updated DataFrame with the 'Read Proportion' and 'Run Pair' columns added.

    Raises:
    ValueError: If any of the pools in the pairs are not found in the DataFrame.

    Example:
    >>> summary_df = pd.DataFrame({
            'Pool': ['SALK082', 'SALK084', 'SALK083', 'SALK085'],
            'Total Reads': [610297, 387352, 500000, 400000]
        })
    >>> updated_df = calculate_read_proportion(summary_df, [('SALK082', 'SALK084'), ('SALK083', 'SALK085')])
    >>> print(updated_df)
           Pool  Total Reads  Read Proportion         Run Pair
    0  SALK082       610297          0.634694  SALK082_SALK084
    1  SALK084       387352          1.575562  SALK084_SALK082
    2  SALK083       500000          0.800000  SALK083_SALK085
    3  SALK085       400000          1.250000  SALK085_SALK083
    """
    # Validate the presence of required columns
    required_columns = ['Pool', 'Total Reads']
    if not all(column in df.columns for column in required_columns):
        raise ValueError(f"The DataFrame must contain the following columns: {required_columns}")
    
    # Validate the presence of specified pools
    for pair in pool_pairs:
        for pool in pair:
            if pool not in df['Pool'].values:
                raise ValueError(f"Pool '{pool}' not found in the DataFrame.")
    
    # Initialize 'Read Proportion' and 'Run Pair' columns
    df['Read Proportion'] = None
    df['Run Pair'] = None
    
    # Calculate and add the proportions for each pair
    for pools in pool_pairs:
        # Extract the total reads for the given pools
        reads_1 = df.loc[df['Pool'] == pools[0], 'Total Reads'].values[0]
        reads_2 = df.loc[df['Pool'] == pools[1], 'Total Reads'].values[0]
        
        # Calculate the proportions
        proportion_1 = reads_2 / reads_1  # Proportion for pools[0]
        proportion_2 = reads_1 / reads_2  # Proportion for pools[1]
        
        # Add the proportions to the respective rows
        df.loc[df['Pool'] == pools[0], 'Read Proportion'] = proportion_1
        df.loc[df['Pool'] == pools[1], 'Read Proportion'] = proportion_2
        
        # Add the run pair to the respective rows
        run_pair_1 = f"{pools[0]}_{pools[1]}"
        run_pair_2 = f"{pools[1]}_{pools[0]}"
        df.loc[df['Pool'] == pools[0], 'Run Pair'] = run_pair_1
        df.loc[df['Pool'] == pools[1], 'Run Pair'] = run_pair_2
    
    return df
```

```python
import sqlite3

# Connect to SQLite database (it will create the database file if it does not exist)
conn = sqlite3.connect('/gale/netapp/seq11/illumina_runs/qc/pools.db')
cursor = conn.cursor()

# Create Pools table
create_pools_table = """
CREATE TABLE IF NOT EXISTS Pools (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT UNIQUE NOT NULL,
    size_bp INTEGER,
    concentration_ng_ul REAL,
    amount_ng REAL
);
"""
cursor.execute(create_pools_table)

# Create Reads table
create_reads_table = """
CREATE TABLE IF NOT EXISTS Reads (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    pool_id INTEGER,
    total_reads INTEGER,
    FOREIGN KEY (pool_id) REFERENCES Pools(id)
);
"""
cursor.execute(create_reads_table)

# Create Proportions table with UNIQUE constraint
create_proportions_table = """
CREATE TABLE IF NOT EXISTS Proportions (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    pool1_id INTEGER,
    pool2_id INTEGER,
    read_proportion REAL,
    run_pair TEXT,
    FOREIGN KEY (pool1_id) REFERENCES Pools(id),
    FOREIGN KEY (pool2_id) REFERENCES Pools(id),
    UNIQUE (pool1_id, pool2_id)
);
"""
cursor.execute(create_proportions_table)

# Commit the changes
conn.commit()

# Close the connection
conn.close()
```

```python
import pandas as pd

def calculate_read_proportion(df, pool_pairs):
    """
    Calculate and add the read proportions and run pairs to the DataFrame for multiple pairs of pools.

    This function takes a DataFrame containing read counts for different pools
    and a list of tuples, each containing two pool names. It calculates the ratio of total reads
    between each specified pair of pools and adds the result to the DataFrame
    in the 'Read Proportion' column for each pool. It also adds a 'Run Pair'
    column to indicate which pools are paired together for sequencing, reflecting
    the order of numerator and denominator in the pair.

    Parameters:
    df (pd.DataFrame): A DataFrame containing at least two columns:
                       'Pool' (str) - the name of the pool
                       'Total Reads' (int) - the total number of reads for the pool.
    pool_pairs (list of tuples): A list of tuples, each containing the names of two pools as strings.
                                 Example: [('SALK082', 'SALK084'), ('SALK083', 'SALK085')]

    Returns:
    pd.DataFrame: The updated DataFrame with the 'Read Proportion' and 'Run Pair' columns added.

    Raises:
    ValueError: If any of the pools in the pairs are not found in the DataFrame.

    Example:
    >>> summary_df = pd.DataFrame({
            'Pool': ['SALK082', 'SALK084', 'SALK083', 'SALK085'],
            'Total Reads': [610297, 387352, 500000, 400000]
        })
    >>> updated_df = calculate_read_proportion(summary_df, [('SALK082', 'SALK084'), ('SALK083', 'SALK085')])
    >>> print(updated_df)
           Pool  Total Reads  Read Proportion         Run Pair
    0  SALK082       610297          0.634694  SALK082_SALK084
    1  SALK084       387352          1.575562  SALK084_SALK082
    2  SALK083       500000          0.800000  SALK083_SALK085
    3  SALK085       400000          1.250000  SALK085_SALK083
    """
    # Validate the presence of required columns
    required_columns = ['Pool', 'Total Reads']
    if not all(column in df.columns for column in required_columns):
        raise ValueError(f"The DataFrame must contain the following columns: {required_columns}")
    
    # Validate the presence of specified pools
    for pair in pool_pairs:
        for pool in pair:
            if pool not in df['Pool'].values:
                raise ValueError(f"Pool '{pool}' not found in the DataFrame.")
    
    # Initialize 'Read Proportion' and 'Run Pair' columns
    df['Read Proportion'] = None
    df['Run Pair'] = None
    
    # Calculate and add the proportions for each pair
    for pools in pool_pairs:
        # Extract the total reads for the given pools
        reads_1 = df.loc[df['Pool'] == pools[0], 'Total Reads'].values[0]
        reads_2 = df.loc[df['Pool'] == pools[1], 'Total Reads'].values[0]
        
        # Calculate the proportions
        proportion_1 = reads_2 / reads_1  # Proportion for pools[0]
        proportion_2 = reads_1 / reads_2  # Proportion for pools[1]
        
        # Add the proportions to the respective rows
        df.loc[df['Pool'] == pools[0], 'Read Proportion'] = proportion_1
        df.loc[df['Pool'] == pools[1], 'Read Proportion'] = proportion_2
        
        # Add the run pair to the respective rows
        run_pair_1 = f"{pools[0]}_{pools[1]}"
        run_pair_2 = f"{pools[1]}_{pools[0]}"
        df.loc[df['Pool'] == pools[0], 'Run Pair'] = run_pair_1
        df.loc[df['Pool'] == pools[1], 'Run Pair'] = run_pair_2
    
    return df
```

```python
import sqlite3

def insert_pools_into_db(df, db_name):
    """
    Insert pool names into the Pools table if they do not already exist.

    Parameters:
    df (pd.DataFrame): DataFrame containing pool names.
    db_name (str): Name of the SQLite database file.
    """
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()
    
    for pool in df['Pool'].unique():
        cursor.execute("""
            INSERT OR IGNORE INTO Pools (name) VALUES (?);
        """, (pool,))
    
    conn.commit()
    conn.close()

def update_database_with_proportions(df, db_name):
    """
    Update the SQLite database with read proportions and run pairs from the DataFrame.

    Parameters:
    df (pd.DataFrame): DataFrame containing updated read proportions and run pairs.
    db_name (str): Name of the SQLite database file.
    """
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()

    for _, row in df.iterrows():
        cursor.execute("SELECT id FROM Pools WHERE name = ?", (row['Pool'],))
        pool1_id_result = cursor.fetchone()
        if pool1_id_result is None:
            print(f"Error: Pool '{row['Pool']}' not found in the database.")
            continue
        pool1_id = pool1_id_result[0]

        run_pair = row['Run Pair'].split('_')
        cursor.execute("SELECT id FROM Pools WHERE name = ?", (run_pair[1],))
        pool2_id_result = cursor.fetchone()
        if pool2_id_result is None:
            print(f"Error: Pool '{run_pair[1]}' not found in the database.")
            continue
        pool2_id = pool2_id_result[0]

        try:
            cursor.execute("""
                INSERT INTO Proportions (pool1_id, pool2_id, read_proportion, run_pair)
                VALUES (?, ?, ?, ?)
            """, (pool1_id, pool2_id, row['Read Proportion'], row['Run Pair']))
        except sqlite3.IntegrityError:
            cursor.execute("""
                UPDATE Proportions
                SET read_proportion = ?, run_pair = ?
                WHERE pool1_id = ? AND pool2_id = ?
            """, (row['Read Proportion'], row['Run Pair'], pool1_id, pool2_id))

    conn.commit()
    conn.close()
```

```python
# Create the DataFrame
summary_df = pd.DataFrame({
    'Pool': ['SALK082', 'SALK084', 'SALK083', 'SALK085'],
    'Total Reads': [610297, 387352, 500000, 400000]
})

# Define the pool pairs
pool_pairs = [('SALK082', 'SALK084'), ('SALK083', 'SALK085')]

# Calculate the read proportions
updated_df = calculate_read_proportion(summary_df, pool_pairs)

# Insert pool names into the database
insert_pools_into_db(summary_df, '/gale/netapp/seq11/illumina_runs/qc/pools.db')

# Update the database with the read proportions
update_database_with_proportions(updated_df, '/gale/netapp/seq11/illumina_runs/qc/pools.db')

```

```python
import sqlite3
import pandas as pd

def fetch_data_from_db(db_name):
    # Connect to the SQLite database
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()
    
    # Fetch data from Pools table
    cursor.execute("SELECT * FROM Pools")
    pools_data = cursor.fetchall()
    pools_df = pd.DataFrame(pools_data, columns=[desc[0] for desc in cursor.description])
    
    # Fetch data from Reads table
    cursor.execute("SELECT * FROM Reads")
    reads_data = cursor.fetchall()
    reads_df = pd.DataFrame(reads_data, columns=[desc[0] for desc in cursor.description])
    
    # Fetch data from Proportions table
    cursor.execute("SELECT * FROM Proportions")
    proportions_data = cursor.fetchall()
    proportions_df = pd.DataFrame(proportions_data, columns=[desc[0] for desc in cursor.description])
    
    # Close the connection
    conn.close()
    
    return pools_df, reads_df, proportions_df

# Example usage
pools_df, reads_df, proportions_df = fetch_data_from_db('/gale/netapp/seq11/illumina_runs/qc/pools.db')

# Display the data
print("Pools Table:")
print(pools_df)

print("\nReads Table:")
print(reads_df)

print("\nProportions Table:")
print(proportions_df)
```

```python
def insert_reads_into_db(df, db_name):
    """
    Insert total reads into the Reads table.

    Parameters:
    df (pd.DataFrame): DataFrame containing pool names and total reads.
    db_name (str): Name of the SQLite database file.
    """
    conn = sqlite3.connect(db_name)
    cursor = conn.cursor()
    
    for _, row in df.iterrows():
        cursor.execute("SELECT id FROM Pools WHERE name = ?", (row['Pool'],))
        pool_id_result = cursor.fetchone()
        if pool_id_result is None:
            print(f"Error: Pool '{row['Pool']}' not found in the database.")
            continue
        pool_id = pool_id_result[0]
        
        cursor.execute("""
            INSERT INTO Reads (pool_id, total_reads)
            VALUES (?, ?)
            ON CONFLICT(pool_id) DO UPDATE SET
                total_reads = excluded.total_reads;
        """, (pool_id, row['Total Reads']))
    
    conn.commit()
    conn.close()
```

```python
def update_database(df, pool_pairs, db_name):
    """
    Update the SQLite database with pool names, total reads, read proportions, and run pairs from the DataFrame.

    Parameters:
    df (pd.DataFrame): DataFrame containing pool names and total reads.
    pool_pairs (list of tuples): A list of tuples, each containing the names of two pools as strings.
    db_name (str): Name of the SQLite database file.
    """
    # Calculate read proportions
    df = calculate_read_proportion(df, pool_pairs)
    
    # Insert pool names into the Pools table
    insert_pools_into_db(df, db_name)
    
    # Insert total reads into the Reads table
    insert_reads_into_db(df, db_name)
    
    # Insert proportions into the Proportions table
    insert_proportions_into_db(df, db_name)
```
