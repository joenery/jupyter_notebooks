```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the sequences
sequences = [
    "ACATTGGCACATTGGC", 
    "CGAACTGTCGAACTGT"
]

# Define the channel nucleotides
green_nucleotides = {'A', 'C'}
red_nucleotides = {'T', 'G'}

# Function to calculate the percentage of green and red nucleotides at each position
def calculate_color_balance(sequences):
    sequence_length = len(sequences[0])
    balance_info = []
    
    for i in range(sequence_length):
        bases_at_position = [sequence[i] for sequence in sequences]
        total_bases = len(bases_at_position)
        green_count = sum(base in green_nucleotides for base in bases_at_position)
        red_count = sum(base in red_nucleotides for base in bases_at_position)
        
        green_percentage = (green_count / total_bases) * 100
        red_percentage = (red_count / total_bases) * 100
        
        balance_info.append({
            "base": i + 1,
            "green_percentage": green_percentage,
            "red_percentage": red_percentage
        })
    
    return balance_info

# Apply the function to the sequences
color_balance = calculate_color_balance(sequences)

# Display the results
for info in color_balance:
    print(f"Base {info['base']}: Green = {info['green_percentage']}%, Red = {info['red_percentage']}%")

# Extract data for plotting
bases = [info['base'] for info in color_balance]
green_percentages = [info['green_percentage'] for info in color_balance]
red_percentages = [info['red_percentage'] for info in color_balance]

# Create stacked bar charts
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.5

bar1 = ax.bar(bases, green_percentages, bar_width, label='Green', color='#66c2a5')
bar2 = ax.bar(bases, red_percentages, bar_width, bottom=green_percentages, label='Red', color='#e41a1c')

ax.set_xlabel('Base Position')
ax.set_ylabel('Percentage')
ax.set_title('Color Balance at Each Base Position')
ax.legend()

plt.xticks(bases)
plt.tight_layout()
plt.show()



```
